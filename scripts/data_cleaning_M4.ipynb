```python
!pip install pandas requests

import pandas as pd
import requests
import os
from datetime import datetime

def import_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        if not os.path.exists("data"):
            os.makedirs("data")
        file_path = "data/dataset_M4.csv"
        with open(file_path, "wb") as f:
            f.write(response.content)
        df = pd.read_csv(file_path)
        return df
    else:
        print("Error downloading dataset.")
        return None

url = "https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&starttime=2023-01-01&endtime=2023-01-02"
df = import_data(url)
df.head()

def clean_data(df):
    if 'time' in df.columns:
        df['time'] = pd.to_datetime(df['time']).dt.strftime('%d-%m-%Y %H:%M:%S')
    if not os.path.exists("output"):
        os.makedirs("output")
    output_path = "output/cleaned_data_M4.csv"
    df.to_csv(output_path, index=False)
    return df

df_cleaned = clean_data(df)
df_cleaned.head()
df_cleaned.info()

```
